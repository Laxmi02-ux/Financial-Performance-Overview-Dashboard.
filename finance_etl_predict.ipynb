```python
import sqlite3
import pandas as pd


conn = sqlite3.connect('finance.db')


try:
    df = pd.read_csv(r'C:\Users\Muskan\OneDrive\Documents\financial_dataset.csv')
    
    #  Clean column names (Remove spaces)
    df.columns = df.columns.str.strip().str.replace(' ', '_')

    df.to_sql('cash_ledger', conn, if_exists='replace', index=False)
    
    print("Success! Your SQLite database now has real Kaggle data.")
    
except FileNotFoundError:
    print(" Error: I couldn't find the CSV file. Make sure it's in the same folder!")

conn.close()
```
